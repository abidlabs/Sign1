```python
from datetime import datetime
from fastbook import *
from fastai.vision.all import *
from fastai.vision.widgets import *
import fastai
import gc
import torch


import numpy as np

nimages_in_sample = 1500
# nimages_in_sample = 100


WITH_MATRIX = True
```

```python
data = 'frank-1-L_culled_100'
# data = 'combined1'
path = f'../data/{data}/Training_Set'
path2 = None
```

```python
image_files = get_image_files(path)

def get_test_fnames(a):
    return random.sample(image_files, nimages_in_sample)
```

```python
# num = 7
low_epochs = 30
num_epochs = 40
num_pred_runs = 5
# RUN_NAME = '20210122-0037 - arch=xresnet50_deep - samples=3000 frozen=1 epochs=40 bs=16 res=360 _data=external'
# RUN_NAME = '20210122-2356 - arch=xresnet50_deep - samples=3000 frozen=1 epochs=40 bs=16 res=360 _data=external'
# RUN_NAME = '20210123-2127 - arch=xresnet50_deep - samples=2000 frozen=1 epochs=22 bs=16 res=360 _data=external'
# RUN_NAME = '20210124-1532 - arch=xresnet50_deep - samples=800 frozen=1 epochs=22 bs=16 res=360 _data=external'
# RUN_NAME = '20210124-2125 - arch=xresnet18 - samples=800 frozen=1 epochs=30 bs=16 res=360 _data=external'
# RUN_NAME = '20210125-1430 - arch=xresnet50 - samples=200 frozen=1 epochs=1 bs=16 res=360 _data=frank'
# RUN_NAME = '20210125-0016 - arch=xresnet50 - samples=800 frozen=1 epochs=20 bs=16 res=360 _data=combined1'
# RUN_NAME = '20210125-1542 - arch=xresnet50 - samples=800 frozen=1 epochs=20 bs=16 res=360 _data=combined1'
# RUN_NAME = '20210126-0012 - arch=xresnet50 - samples=1000 frozen=1 epochs=30 bs=16 res=360 _data=combined1'
# RUN_NAME = '20210126-0759 - arch=xresnet50_deep - samples=1000 frozen=1 epochs=30 bs=16 res=360 _data=combined1'
# RUN_NAME = '20210126-1840 - arch=xresnet50_deep - samples=1000 frozen=1 epochs=30 bs=16 res=360 _data=combined1'
# RUN_NAME = '20210127-1717 - arch=xresnet50 - samples=1100 frozen=1 epochs=20 bs=28 res=280 _data=combined2'
RUN_NAME = '20210127-2251 - arch=xresnet50 - samples=1100 frozen=1 epochs=40 bs=28 res=280 _data=combined2'


RESOLUTION = 280
BATCH_SIZE=28

accd = dict()

test_learn = load_learner(f'../models/{RUN_NAME}.pkl', cpu=False)

for ep in range(low_epochs, num_epochs):
    acc_list = []
    print(f'\n\nEPOCH #{ep}')
    
    for pr in range(1, num_pred_runs+1):

        load_model(f'models/{RUN_NAME}_{ep}.pth', test_learn, opt=Adam, with_opt=False)



        test_db = DataBlock(
            blocks=(ImageBlock, CategoryBlock), 
            get_items=get_test_fnames,
            get_y=parent_label, 
            item_tfms=Resize(RESOLUTION, method='bilinear')) #,
#             item_tfms=CropPad(RESOLUTION, pad_mode='zeros'))
#            ,    batch_tfms=aug_transforms(do_flip=True, size=RESOLUTION, batch=False, max_zoom=1.0, mult=1, pad_mode='zeros'))


        dls = test_db.dataloaders(path, bs=BATCH_SIZE)

        test_dl = dls.test_dl(get_test_fnames(''), with_labels=True)

        with test_learn.no_bar():
            preds, targs = test_learn.get_preds(dl=test_dl)

        acc_list.append(accuracy(preds, targs))
        print(f'pred_run: {pr}, accuracy: {round(np.float(100* accuracy(preds, targs)))}%')
    
    accd[ep] = np.mean(acc_list)
    print(f'AVERAGE ACCURACY:      {round(100*np.mean(acc_list),2)}%')
    
    if WITH_MATRIX:
        interp = ClassificationInterpretation.from_learner(test_learn, dl=test_dl)
        interp.plot_confusion_matrix(figsize=(7,7))
        plt.show()
    

```

```python
print('\n   Best Run ')
print('--------------')
print(f'Epoch #:   {max(accd, key=accd.get)}\nAccuracy:  {round(100*max(accd.values()),2)}%')
```
